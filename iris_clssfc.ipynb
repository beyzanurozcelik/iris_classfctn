```python
import pandas as pd
```

```python
data = pd.read_csv("IRIS.csv")
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder

data["species"] = LabelEncoder().fit_transform(data["species"].values)
data.head()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop("species",axis=1)
y = data["species"]

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.3,random_state=101)

x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter=150)
```

```python
model.fit(x_train,y_train)

predicted = model.predict(x_test)
predicted.size
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,predicted)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predicted))
```

```python
from sklearn.svm import LinearSVC

model2 = LinearSVC()

model2.fit(x_train,y_train)

predicted2 = model2.predict(x_test)

predicted2.size
```

```python
confusion_matrix(y_test,predicted2)
```

```python
print(classification_report(y_test,predicted2))
```
